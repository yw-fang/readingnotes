```python
%matplotlib inline

# For the reference of the default color map of matplotlib, see
# https://matplotlib.org/users/dflt_style_changes.html

#For the fonts in matplotlib, please see
#https://matplotlib.org/users/dflt_style_changes.html#figure-size-font-size-and-screen-dpi
# This example is from CHAPTER 3 of the book "
# matplotlib for developpers",
# but I made many revisions

import matplotlib.style
import matplotlib as mpl
mpl.style.use('classic')
mpl.rcParams['figure.facecolor'] = '1'
#if choose the grey backgroud, use 0.75
mpl.rcParams['figure.figsize'] = [8,12.5]
mpl.rcParams['lines.linewidth'] = 1.5
mpl.rcParams['legend.fancybox'] = True
mpl.rcParams['legend.fontsize'] = 19
mpl.rcParams['legend.scatterpoints'] = 1 #scatterpoints,
#it's the numer of maker points in the legend when
#creating a legend entry for a scatter plot
mpl.rcParams["axes.formatter.useoffset"]=False #turn off the axis offset-values. 
# If on. the axis label will use a offset value by the side of axis
mpl.rcParams["axes.linewidth"]  = 1.5 #change the boarder width
#plt.rcParams["axes.edgecolor"] = "0.15" #change the boarder color

ticklabel_size = 19
mpl.rcParams['xtick.labelsize'] = ticklabel_size
mpl.rcParams['ytick.labelsize'] = ticklabel_size

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pandas import DataFrame

#import pdf module
from matplotlib.backends.backend_pdf import PdfPages

data=pd.read_csv('./sDFT-SOC-U-verification/type4-U-test.txt', delim_whitespace=True)
data1=pd.read_csv('./sDFT-SOC-U-verification/type4-U-SOC-test.txt', delim_whitespace=True)



frame = DataFrame(data)
frame1 = DataFrame(data1)
#frame['${E_g}$']=frame['${\Delta}$'].values


fig, axes = plt.subplots(nrows=2, ncols=1)

with PdfPages('./phase-diagram-sPBEUSOC.pdf') as pdf:
    ax1=frame.plot(['$U$'],secondary_y=['${E_g}$'],color={'r','b'},ax=axes[0],
                  kind='line', marker='o',markersize=10,mark_right=False)
    ax1.set_ylabel('Magnetic moment  $M$ of Os ($\mu_B$)',fontsize=22)
    ax1.set_xlabel('$U$ (eV)', fontsize = 22)
    ax1.right_ax.set_ylabel('Band gap  ${E_g}$ (eV)',fontsize=22)
    #ax1.annotate('', (1.5, 5.2), (0, 5.2), arrowprops={'arrowstyle':'<-',"facecolor": 'm'})
    ax1.annotate('', (1, 0.6), (1.5, 0.6), arrowprops=dict(facecolor='b',shrink=0.1)) # arrowstyle="<-"
    ax1.annotate('', (4, 0.6), (3.5, 0.6), arrowprops=dict(facecolor='r',shrink=0.1)) # arrowstyle="<-"
    ax1.text(-0.7, 1.78, '(a)', fontsize=22)
    ax1.text(0.5, 1.2, 'PBEsol+$U$', fontsize=22,color='k')
    plt.xticks(np.arange(0, 5, 1))
    plt.tight_layout()
    ax1.grid()
    
    
    ax1=frame1.plot(['$U$'],secondary_y=['${E_g}$'],color={'b','r'},ax=axes[1],
                   kind='line', marker='o',markersize=10,mark_right=False,
                   legend=False)
    ax1.set_ylabel('Magnetic moment  $M$ of Os ($\mu_B$)',fontsize=22)
    ax1.set_xlabel('$U$ (eV)', fontsize = 22)
    ax1.right_ax.set_ylabel('Band gap ${E_g}$ (eV)',fontsize=22)
    #ax1.annotate('', (1.5, 5.2), (0, 5.2), arrowprops={'arrowstyle':'<-',"facecolor": 'm'})
    ax1.annotate('', (1, 0.6), (1.5, 0.6), arrowprops=dict(facecolor='b',shrink=0.1)) # arrowstyle="<-"
    ax1.annotate('', (4, 0.6), (3.5, 0.6), arrowprops=dict(facecolor='r',shrink=0.1)) # arrowstyle="<-"
    ax1.text(-0.7, 1.78, '(b)', fontsize=22)
    ax1.text(0.5, 1.2, 'PBEsol+SOC+$U$', fontsize=22,color='k')
    ax1.set_ylim([0.0,1.8])
    ax1.right_ax.set_ylim([0.0,0.40])
    plt.xticks(np.arange(0, 5, 1))
    plt.tight_layout()
    ax1.grid()

    pdf.savefig()
    plt.close()
```

A comment to my code about DataFrame plot:
    
    when we want to plot several plots in one PDF using DataFrame plot, we can use this command to add subplot:
    fig, axes = plt.subplots(nrows=2, ncols=1)
    
    in addition, ax must be specified after x and y. for example: frame1.plot(['$U$'],secondary_y=['${E_g}$'],color={'b','r'},ax=axes[1],ax=axes[1]) 
    
    It should be mentioned that we cannot put ax before x and y variables we want to plot.
