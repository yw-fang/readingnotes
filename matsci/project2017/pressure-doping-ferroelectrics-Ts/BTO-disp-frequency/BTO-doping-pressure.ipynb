```python
%matplotlib inline

# For the reference of the default color map of matplotlib, see
# https://matplotlib.org/users/dflt_style_changes.html

#For the fonts in matplotlib, please see
#https://matplotlib.org/users/dflt_style_changes.html#figure-size-font-size-and-screen-dpi
# This example is from CHAPTER 3 of the book "
# matplotlib for developpers",
# but I made many revisions

import matplotlib.style
import matplotlib as mpl
mpl.style.use('classic')
mpl.rcParams['figure.facecolor'] = '1'
#if choose the grey backgroud, use 0.75
mpl.rcParams['figure.figsize'] = [9,6]
mpl.rcParams['lines.linewidth'] = 2.2
mpl.rcParams['legend.fancybox'] = True
mpl.rcParams['legend.fontsize'] = 19
mpl.rcParams['legend.scatterpoints'] = 1 #scatterpoints,
#it's the numer of maker points in the legend when
#creating a legend entry for a scatter plot
mpl.rcParams["axes.formatter.useoffset"]=False #turn off the axis offset-values. 
# If on. the axis label will use a offset value by the side of axis
mpl.rcParams["axes.linewidth"]  = 1.5 #change the boarder width
#plt.rcParams["axes.edgecolor"] = "0.15" #change the boarder color

ticklabel_size = 19
mpl.rcParams['xtick.labelsize'] = ticklabel_size
mpl.rcParams['ytick.labelsize'] = ticklabel_size

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pandas import DataFrame

#import pdf module
from matplotlib.backends.backend_pdf import PdfPages

data=pd.read_csv('./BTO-0.05e.csv')

frame = DataFrame(data)

with PdfPages('./BTO-0.05e.pdf') as pdf:
    ax1=frame.plot(['Pkbar'],secondary_y=['TiOP4mm','BiOP4mm'],color={'r','g','b'},
                  kind='line', marker='o',markersize=10,mark_right=False)
    
    ax1.set_ylabel('$\Gamma$-frequency (THz) of 'r'${P4mm}$'' BaTiO$_3$',fontsize=22)
    ax1.set_xlabel('$P$ (GPa)', fontsize = 22)
    
    ax1.set_ylim([0,0.5])
    ax1.yaxis.set_major_formatter(plt.FuncFormatter('{:.1f}$i$'.format)) #set the display format
    #ax1.right_ax.set_ylabel('r$\Gamma$-frequency of ${R\bar{3}c}$ LiOsO$_3$',fontsize=16)
    #ax1.right_ax.set_ylabel('bond angle ($^\circ$) of ${R3c}$ LiOsO$_3$',fontsize=22)
    #ax1.annotate('', (1.5, 5.2), (0, 5.2), arrowprops={'arrowstyle':'<-',"facecolor": 'm'})
    ax1.annotate('', (0, 5.2), (1.5, 5.2), arrowprops=dict(facecolor='r',shrink=0.1)) # arrowstyle="<-"
    ax1.annotate('', (8, 5.55), (6.5, 5.55), arrowprops=dict(facecolor='b',shrink=0.1)) # arrowstyle="<-"
#    plt.xticks(np.arange(0, 13, 2))
    plt.tight_layout()
    ax1.grid()
#    plt.axvline(x=5, color='k', linestyle='--') #add a vertical line
    pdf.savefig()
    plt.close()
```

```python
%matplotlib inline

# For the reference of the default color map of matplotlib, see
# https://matplotlib.org/users/dflt_style_changes.html

#For the fonts in matplotlib, please see
#https://matplotlib.org/users/dflt_style_changes.html#figure-size-font-size-and-screen-dpi
# This example is from CHAPTER 3 of the book "
# matplotlib for developpers",
# but I made many revisions

import matplotlib.style
import matplotlib as mpl
mpl.style.use('classic')
mpl.rcParams['figure.facecolor'] = '1'
#if choose the grey backgroud, use 0.75
mpl.rcParams['figure.figsize'] = [9,6]
mpl.rcParams['lines.linewidth'] = 2.2
mpl.rcParams['legend.fancybox'] = True
mpl.rcParams['legend.fontsize'] = 19
mpl.rcParams['legend.scatterpoints'] = 1 #scatterpoints,
#it's the numer of maker points in the legend when
#creating a legend entry for a scatter plot
mpl.rcParams["axes.formatter.useoffset"]=False #turn off the axis offset-values. 
# If on. the axis label will use a offset value by the side of axis
mpl.rcParams["axes.linewidth"]  = 1.5 #change the boarder width
#plt.rcParams["axes.edgecolor"] = "0.15" #change the boarder color

ticklabel_size = 19
mpl.rcParams['xtick.labelsize'] = ticklabel_size
mpl.rcParams['ytick.labelsize'] = ticklabel_size

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pandas import DataFrame

data=pd.read_csv('./BTO-0.05e.csv')

frame = DataFrame(data)

fig = plt.figure() # Create matplotlib figure

ax = fig.add_subplot(111) # Create matplotlib axes
ax2 = ax.twinx() # Create another axes that shares the same x-axis as ax.

df=frame.copy()
df['Pkbar']=df['Pkbar']/10 #conver kbar to GPa
df.set_index('Pkbar', inplace=True)
df.TiOP4mm.plot(ax=ax,color={'r'},alpha=0.9,
                  kind='line', marker='o',markersize=12,mark_right=False, label='Ti-O')
df.BaOP4mm.plot(ax=ax,color={'b'},alpha=0.9,
                  kind='line', marker='o',markersize=12,mark_right=False, label='Ba-O')
ax.legend(loc="upper center")
df.PEfrequency.plot(ax=ax2,color={'g'},alpha=0.9,
                  kind='line', marker='s',markersize=10,mark_right=False, label='$\Gamma$')
#ax.legend(loc="best")

ax.set_xlabel('$P$ (GPa)', fontsize = 20)
ax.set_ylabel(r'$M$-O displacements (${\mathrm{\AA}}$)', fontsize = 20)
ax2.set_ylabel('$\Gamma$-frequency (THz)', fontsize = 20)


ax.annotate('', (1.1, 0.095), (1.8, 0.095), arrowprops=dict(facecolor='r',shrink=0.1)) # arrowstyle="<-"
ax.annotate('', (1.1, 0.065), (1.8, 0.065), arrowprops=dict(facecolor='b',shrink=0.1)) # arrowstyle="<-"
ax.annotate('', (9.8, 0.065), (9.1, 0.065), arrowprops=dict(facecolor='g',shrink=0.1)) # arrowstyle="<-"
#    plt.xticks(np.arange(0, 13, 2))
plt.tight_layout()
ax.grid()

#plt.show()
plt.savefig('BTO-0.05e.pdf')
```
